## Challenge 1: Word Frequency Analyzer
Write a function `top_k_words(filename: str, k: int) -> list` that returns the top k frequent words.
- Ignore case and punctuation.
- Sort by frequency, then alphabetically.


```python
import re
from collections import Counter

def top_k_words(filename: str, k: int) -> list:
    # Your code here
    with open(filename, "r") as f:
        text = f.read()

    # Convert to lowercase
    text = text.lower()

    # Remove punctuation using regex
    words = re.findall(r'\b\w+\b', text)

    # Count word frequency
    freq = Counter(words)

    # Sort by frequency (descending) and alphabetically (ascending)
    sorted_words = sorted(freq.items(), key=lambda x: (-x[1], x[0]))

    # Return top k words
    return [word for word, count in sorted_words[:k]]



# Sample Test (create a sample file)
with open("sample.txt", "w") as f:
    f.write("Data science is evolving. Python is versatile. Learning never stops.")

print(top_k_words("sample.txt", 3))           #expected ["is", "data", "science"]
```

## Challenge 2 : Analyze daily expenses and extract insights.
User enters:

Number of transactions
Category and amount

Food 200
Travel 500
Food 300
Shopping 700

### Output
* Total expense
* Highest expense category
* Category-wise total

```python
def analyze_expenses():
    n = int(input("Enter number of transactions: "))

    expenses = {}

    for _ in range(n):
        category, amount = input().split()
        amount = float(amount)

        # store category-wise total
        if category in expenses:
            expenses[category] += amount
        else:
            expenses[category] = amount

    # calculate total expense
    total = sum(expenses.values())

    # find highest  category
    highest_category = max(expenses, key=expenses.get)

    print("Total Expense:", total)
    print("Highest Category:", highest_category)
    print("Category-wise:", expenses)


analyze_expenses()
```

### Challenge 3: Data Cleaner (Mini Pandas Logic Without Pandas)

10 ,20 ,abc ,30 ,-5 , 40

* Remove invalid values
* Remove negative values


Show:
* Clean list
* Average
* Max value

```python
def clean_data():
    n = int(input("Enter number of values: "))

    data = []

    for _ in range(n):
        value = input()

        try:
            num = float(value)

            # ignore negative values
            if num >= 0:
                data.append(num)

        except:
            # handle invalid values (ignore them)
            continue

    # calculate average
    if len(data) > 0:
        avg = sum(data) / len(data)
        max_val = max(data)
    else:
        avg = 0
        max_val = None

    print("Clean Data:", data)
    print("Average:", avg)
    print("Max:", max_val)


clean_data()
```
